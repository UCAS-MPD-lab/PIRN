```python
import fluidfoam as ff
import joblib
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
from scipy.interpolate import Rbf
```

```python
labels = np.arange(0.5,2.52,0.02)
u0_all = []
u1_all = []
```

```python
for label in labels:
    path = f"/home/dell/OpenFOAM/dell-v2206/run/cavity4Res/U_series/cavity_{label:0.2f}"
    U = ff.readof.readfield(path, time_name='latestTime', name='U')
    x, y, z = ff.readof.readmesh(path)
    points = np.column_stack((x, y))
    X = np.linspace(0, 0.1, 64)
    Y = np.linspace(0, 0.1, 64)
    xx, yy = np.meshgrid(X, Y)
    points2inter = np.vstack([xx.ravel(), yy.ravel()]).T
    u0_inter = griddata(points, U[0], points2inter, method='linear').reshape(xx.shape)
    u1_inter = griddata(points, U[1], points2inter, method='linear').reshape(xx.shape)
    u0_all.append(u0_inter)
    u1_all.append(u1_inter)
```

```python
for label in labels:
    path = f"/home/dell/OpenFOAM/dell-v2206/run/cavity4Res/U_series/cavity_{label:0.2f}"
    U = ff.readof.readfield(path, time_name='latestTime', name='U')
    x, y, z = ff.readof.readmesh(path)
    rbf0 = Rbf(x, y, U[0], function='linear', smooth=0)
    rbf1 = Rbf(x, y, U[1], function='linear', smooth=0)
    X = np.linspace(0, 0.1, 64)
    Y = np.linspace(0, 0.1, 64)
    xx, yy = np.meshgrid(X, Y)
    u0_inter = rbf0(xx, yy)
    u1_inter = rbf1(xx, yy)
    u0_all.append(u0_inter)
    u1_all.append(u1_inter)
```

```python
Re = 10 * labels
Re = Re.reshape(101, 1)
u0_all = np.array(u0_all).reshape(101, 64, 64, 1)
u1_all = np.array(u1_all).reshape(101, 64, 64, 1)
```

```python
joblib.dump(Re,'Re.joblib')
joblib.dump(u0_all, 'u0_all.joblib')
joblib.dump(u1_all, 'u1_all.joblib')
joblib.dump(xx, 'xx.joblib')
joblib.dump(yy, 'yy.joblib')
```
