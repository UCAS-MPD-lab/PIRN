```python
from Networks.MLP_Dict import MLP_Dict
from Saving.saveFile import get_next_model_filename
from Trainers import split_data,Trainer
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import joblib
```

```python
# 初始化模型
input_shape = (1,)  # 输入形状
output_shape = (1024, )

model = MLP_Dict(input_shape, output_shape, hidden_layers=[16, 32, 128], activation_function="tanh", output_keys=["u"])

# # 打印模型结构
# print(model)
```

```python
input_np = joblib.load('preData/input_data.joblib').astype(np.float32)
u_train_np = joblib.load('preData/true_output.joblib').squeeze(3).astype(np.float32)

input_train = torch.from_numpy(input_np).unsqueeze(1)
u_train = torch.from_numpy(u_train_np).view(100,-1).unsqueeze(1)

targets_train = {"u": u_train}
```

```python
# 定义损失函数和优化器
criterions = {
    "u": nn.MSELoss(),  # 针对 u 的损失函数
}
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
# 导入数据
train_loader, val_loader = split_data(input_train, targets_train, batch_size=256)
trainer = Trainer(model, optimizer, criterions, device="cuda")
```

```python
# 训练
save_path = get_next_model_filename(prefix='MLP')
trainer.fit(train_loader, val_loader, num_epochs=100000, early_stop_patience=1000, save_path=save_path)
```

```python
trainer.plot_loss()
```
